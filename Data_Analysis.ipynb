# Date when most number of questions were asked anonymously
First would be to create a function to convert it into date and after that we will use groupby with count in aggregation with sort also

```python
# So the first task would be to create dates from the given creationDate
# So lets write a function for it
import time
def create_date(cols):
    date = str(cols[0])
    date = str(date[:-3])
    output = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(int(date)))
    date_output = output.split()
    
    return date_output[0]  
```

```python
questions['date'] = questions[['creationDate']].apply(create_date, axis = 1)
questions.head(10)
```

```python
questions.tail()
```

```python
output = questions[questions.isAnonymous == True][['_id', 'date']].groupby(['date'])['_id'].count().reset_index(name='count').sort_values(['count'], ascending=False)
output
```

# Percentage of questions answered after 5 minutes of creation
So the primary focus in this will beto find time after which question is answered of which first the function is made after that we will find which have difference greater than 5 minutes

```python
import datetime as dt
from datetime import timedelta
import time
def answered_in(cols):
    creation_date = str(cols[0])
    creation_date = creation_date[:-3]
    answered_date = str(cols[1])
    answered_date = answered_date[:-3]
    
    start = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(int(creation_date)))
    end = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(int(answered_date)))
    start_dt = dt.datetime.strptime(start, '%Y-%m-%d %H:%M:%S')
    end_dt = dt.datetime.strptime(end, '%Y-%m-%d %H:%M:%S')
    diff = (end_dt - start_dt)

    return diff
```

```python
output = pd.merge(answers, questions, how='left', left_on='parentId', right_on='_id')
output.head()
```

```python
output['answered_after'] = output[['creationDate_y', 'creationDate_x']].apply(answered_in, axis = 1)
output.head()
```

```python
output[output['answered_after'] >= timedelta(seconds=300)].count()
```
